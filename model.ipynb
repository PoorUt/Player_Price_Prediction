```python
df=df.drop(['full_name','b_day','height','weight', 'college'],axis=1)
df.head()
```

```python
df.shape
```

```python
#Lets remove the # from the jersey column

def removehash(value):
    value = value[1:]
    return int(value)



```

```python
df['jersey']=df['jersey'].apply(removehash)
```

```python
df.head()
```

```python
#the salary column has a $ sign at the begining. lets remove that

df['salary']=df['salary'].apply(removehash)
```

```python
df.head()
```

```python
df['team'].value_counts()
```

```python
df['country'].value_counts()
```

```python
#for countries that have only 1 team, we can put them into others
def removecoutryoutlier(value):
    if value not in['USA', 'Canada', 'Australia']:
        return 'Others'
    else:
        return value
    
```

```python
df['country']=df['country'].apply(removecoutryoutlier)
df['country'].value_counts()
```

```python
df.head()
```

```python
df['position'].value_counts()
```

```python
df['draft_round'].value_counts()
```

```python
df['draft_peak'].value_counts()
```

```python
def removeundrafted(value):
    if value=="Undrafted":
        return
    else:
        return value


df['draft_round'] = df['draft_round'].apply(removeundrafted)
df['draft_peak'] = df['draft_peak'].apply(removeundrafted)
```

```python
df.head()
```

```python
df['draft_peak'].value_counts()
```

```python
#lets remove null values where ever there was undrafted.

df=df.dropna()
```

```python
df.head()
```

```python
#Label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['position']=le.fit_transform(df['position'])
df['country']=le.fit_transform(df['country'])
df['team']=le.fit_transform(df['team'])

df.head()
```

```python
df.describe()
```

```python
#Lets create x and y

x = df.drop(['salary',''], axis = 1)
y = df['salary']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
predictions = lr.predict(x_test)
```
